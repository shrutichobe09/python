Q1) What are the pros and cons of a microservices architecture?

Ans- Microservices architecture is nothing but one application or
software is broken down into smaller part or module and that small part
of the application will work a particular task assigned to it and by
combining all the microservices we can make application each
microservice is responsible to do particular job and if there is any
error in the application so we can focus on particular microservice
easily so whole application will not go down Each microservice can be
deployed independently, allowing for faster development, Microservices
should communicate through well-defined APIs,

Advantages of Microservices: i) Microservices can be easily scalable we
can scale up and scale down according to our requirement ii) If an error
occurs in one microservice, it will not affect the entire system.
iii)microservices can be deployed independently, so the speed of the
application will be fast

Disadvantages of microservices i) Breaking down the application into the
smaller part of microservices it can lead to complexity of the
architecture and combining all the microservices will be the complex
task ii) Testing and debugging of the application will be difficult
first we need to test the each microservice separately and then by
combining all together we need to test it again as a whole application
iii) It will also lead to latency issue iv) testing every part of the
system will increase the cost and effort

1.  What is the difference between is and == in Python when comparing
    variables?

== (Equal Operator):

The == operator checks if the values of two variables are the same. It
compares the data or content of the objects, not their memory locations.

Example:

a = 6 b = 6 print(a == b) \# True (values are the same so it will return
true)

is (Identity Operator):

The is operator checks if two variables refer to the same object in
memory means if they are the same object, not just equal in value It
compares the memory address of the objects, not their contents.

Example:

a = \[1, 2, 3\] b = \[1, 2, 3\] print(a is b) \# False

1.  How does Python handle variable assignment and memory management?
    When you assign a value to a variable, Python doesn’t store the
    value directly in the variable. it creates an object in memory and
    binds the variable to that object. Python creates an integer object
    10 in memory and makes x point to that object.

example:

x = 10 if you assign the same value to another variable, both variables
point to the same object in memory: y = 10 both x and y point to the
same 10 object in memory.

Python automatically handles memory using this two method 1.Reference
Counting Every object in Python has a reference count, which tracks how
many variables point to the object. When an object is no longer needed
Python can safely remove it from memory. and it will free that space for
other variable to store

example: a = \[1, 2, 3\] b = a  
del a \# we will store a in b variable and we will delete a because it
will utilize the unnecessary space

2.Garbage collection It’s like a cleaning the code that automatically
checks for objects that are no longer in use even if they refer to each
other and gets rid of them to free up memory. Python makes sure your
program doesn’t use up too much memory by holding on to unnecessary
objects.

1.  What is the difference between mutable and immutable types in
    Python? Mutable type : Mutable types are objects whose content can
    be changed after the object is created.

Examples: Lists \[12, 25, 33\] Dictionaries {‘a’: 1, ‘b’: 2} Sets {14,
26, 38}

When you change a mutable object (like adding or removing items from a
list), you’re changing the same object in memory. No new object is
created.

Example my_list = \[21, 25, 33\] \# Create a list my_list.append(46) \#
Modify the same list by adding a new element print(my_list) \# Output:
\[21, 25, 33, 46\]

Immutable Type: Immutable types are objects whose content cannot be
changed after the object is created. If you try to modify them, Python
creates a new object instead.

Examples:

Integers :10 Strings :“hello” Tuples :1, 2, 3

If you try to change an immutable object (like changing a string),
Python creates a new object with the new value and keeps the original
object unchanged. Example:

x = 10 x = x + 1 \# This creates a new integer object (11) the original
10 is unchanged print(x) \# it will print 11

1.  What is the difference between random.sample() and random.shuffle()?

<!-- -->

1.  random.sample() it is used to select a specified number of unique
    items from a sequence without modifying the original sequence. It
    doesn’t change the original list. we can specify the number of items
    you want to select. It returns a new list of sampled elements.

example: import random my_list = \[1, 2, 3, 4, 5\]

# it will take 3 random element from list

sampled = random.sample(my_list, 3) \# it will Return a new list with 3
random items print(“Random sample:”, sampled)

# The original list my_list will remain unchanged

print(“Original list after sample:”, my_list) \# Output: \[1, 2, 3, 4,
5\]

1.  random.shuffle(): it is used to randomly shuffle all the items in a
    sequence it modifies the original list itself. It doesn’t return
    anything (it modifies the list directly) It doesn’t allow you to
    specify how many elements to shuffle; it shuffles the entire list.

example: my_list = \[1, 2, 3, 4, 5\] random.shuffle(my_list) \# This
will shuffle the entire list print(“List after shuffle:”, my_list) \#
Output: \[3, 5, 1, 2, 4\] (elements order can be anything)

1.  What is the purpose of the not operator in Python?

The not operator in Python is a logical negation operator. It is used to
invert the boolean value of a condition. If the condition is True, not
makes it False; if the condition is False, not makes it True.

Purpose: Negates a boolean expression. Useful for checking the opposite
of a condition.

example: x = True y = False

# Using the ‘not’ operator

print(not x) \# False, because x is True print(not y) \# True, because y
is False

age = 20

if not age \< 18: print(“You are an adult eligble for voting”)
